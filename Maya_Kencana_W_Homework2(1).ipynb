```python
#Soalnomor1
ba = int(input("masukkan ba: "))
bw = int(input("masukkan bw: "))

while(ba != bw):
    if(ba > bw):
        if(ba != bw + 1):
            ba = ba - 1
            print(ba)
        else:
            ba = ba - 1
    else:
        if(ba != bw - 1):
            ba = ba + 1
            print(ba)
        else:
            ba = ba + 1
  
  
```

```python
#Soalnomor2
#Rumus Bunga Majemuk Mt=Mo*(1+sukubunga)^periode
na = float(input("nilai awal investasi: "))
li = float(input("lama investasi: "))
sb = float(input("suku bunga: "))
i = 1

while(i != li + 1):
    na =(na*((1+sb/100)**i))
    print("tahun ke", i, "jumlah investasi anda", na)
    i = i + 1
```
